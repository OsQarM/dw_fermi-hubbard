```python
import sys, os
sys.path.append(os.path.abspath('..'))

import numpy as np
import matplotlib.pyplot as plt

from models.qutip_model import QutipFermiHubbard
from models.qilisdk_model import QilisdkFermiHubbard
```

```python
backend = 'qilisdk'  # 'qutip' or 'qilisdk'

chain_length = 5
gx = 1
gz = 10
e = 2*gz
T = 5.0

if backend == 'qutip':
    nsteps = int(T)*100
    model = QutipFermiHubbard(chain_length, gx, gz, e, T, nsteps)
elif backend == 'qilisdk':
    dt = 0.01
    model = QilisdkFermiHubbard(chain_length, gx, gz, e, T, dt)
else:
    raise ValueError(f'Unknown backend: {backend}')

expectation_values = model.run()
```

```python
data_array = np.array(expectation_values)

plt.figure(figsize=(10, 6))

n_qubits = data_array.shape[1]
time_steps = np.arange(data_array.shape[0])

half = n_qubits // 2
for i in range(half):
    plt.plot(time_steps, data_array[:, i], label=f'Upper {i}', marker='o', markersize=4)
for i in range(half, n_qubits):
    plt.plot(time_steps, data_array[:, i], label=f'Lower {i-half}', marker='s', markersize=4)

plt.title('Evolution of Qubit Expectation Values')
plt.xlabel('Time Step')
plt.ylabel('Expectation Value')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()

plt.show()
```
